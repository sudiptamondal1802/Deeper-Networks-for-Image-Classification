```python
import pickle
import os
import matplotlib.pyplot as plt
```

```python
SAVE_PATH = '/Users/sudiptamondal/Documents/QMUL/Deep Learning with Computer Vision/DL CV Project/PICKLE_DUMP'
```

```python
def load_pickle_file(model):
    my_file = open(os.path.join(SAVE_PATH,model),'rb')
    return(pickle.load(my_file))
```

### Loss and accuracy plot for CIFAR10 baseline models

```python
cifar10_googlenet_model1 = load_pickle_file('cifar10_googlenet_model1')
cifar10_vgg19_model1     = load_pickle_file('cifar10_vgg19_model1')
cifar10_resnet18_model1  = load_pickle_file('cifar10_resnet18_model1')
cifar10_resnet34_model1  = load_pickle_file('cifar10_resnet34_model1')
cifar10_resnet50_model1  = load_pickle_file('cifar10_resnet50_model1')
```

```python
""" Evaluating the all model performance for the train losses"""
plt.style.use('ggplot')
plt.plot(cifar10_googlenet_model1['train_losses'], label='cifar10_googlenet_model1',color='blue')
plt.plot(cifar10_vgg19_model1['train_losses'], label='cifar10_vgg19_model1',color='green')
plt.plot(cifar10_resnet18_model1['train_losses'], label='cifar10_resnet18_model1',color='red')
plt.plot(cifar10_resnet34_model1['train_losses'], label='cifar10_resnet34_model1',color='cyan')
plt.plot(cifar10_resnet50_model1['train_losses'], label='cifar10_resnet50_model1',color='purple')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.tight_layout()
plt.savefig(os.path.join(SAVE_PATH,'cifar10_train_losses.png'))
plt.show()
plt.close()
```

```python
""" Evaluating the all model performance for the validation losses"""
plt.style.use('ggplot')
plt.plot(cifar10_googlenet_model1['valid_losses'], label='cifar10_googlenet_model1',color='blue')
plt.plot(cifar10_vgg19_model1['valid_losses'], label='cifar10_vgg19_model1',color='green')
plt.plot(cifar10_resnet18_model1['valid_losses'], label='cifar10_resnet18_model1',color='red')
plt.plot(cifar10_resnet34_model1['valid_losses'], label='cifar10_resnet34_model1',color='cyan')
plt.plot(cifar10_resnet50_model1['valid_losses'], label='cifar10_resnet50_model1',color='purple')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.tight_layout()
plt.savefig(os.path.join(SAVE_PATH,'cifar10_valid_losses.png'))
plt.show()
plt.close()
```

```python
""" Evaluating the all model performance for the train accuracies"""
plt.style.use('ggplot')
plt.plot(cifar10_googlenet_model1['train_accuracies'], label='cifar10_googlenet_model1',color='blue')
plt.plot(cifar10_vgg19_model1['train_accuracies'], label='cifar10_vgg19_model1',color='green')
plt.plot(cifar10_resnet18_model1['train_accuracies'], label='cifar10_resnet18_model1',color='red')
plt.plot(cifar10_resnet34_model1['train_accuracies'], label='cifar10_resnet34_model1',color='cyan')
plt.plot(cifar10_resnet50_model1['train_accuracies'], label='cifar10_resnet50_model1',color='purple')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.savefig(os.path.join(SAVE_PATH,'cifar10_train_accuracies.png'))
plt.show()
plt.close()
```
